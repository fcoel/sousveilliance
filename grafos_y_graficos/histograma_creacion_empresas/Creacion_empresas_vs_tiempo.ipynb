```python
import datetime

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
path = "../../datos/datos_procesados/fecha_creacion.csv"
df = pd.read_csv(path)

# De float a str
df['fecha_creacion'] = df['fecha_creacion'].astype(str)
# De str a datetime
df['fecha_creacion'] = pd.to_datetime(df['fecha_creacion'], 
                                      format='%Y%m%d', errors='coerce')
df.head()
```

```python
df2 = df
df2.groupby([df2.fecha_creacion.dt.year, df2.fecha_creacion.dt.month]).count().plot(kind="bar")

plt.show()
```
