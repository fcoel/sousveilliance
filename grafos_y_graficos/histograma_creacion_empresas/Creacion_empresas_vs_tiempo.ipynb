```python
import datetime

import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = "../datos/fecha_creacion.csv"
df = pd.read_csv(path)
```

```python
# Parseo a datetime object
def to_datetime(val):
    date = datetime.datetime.strptime(str(val), '%Y%m%d')
    return date

df.fecha_creacion = df.fecha_creacion.apply(to_datetime)
```
