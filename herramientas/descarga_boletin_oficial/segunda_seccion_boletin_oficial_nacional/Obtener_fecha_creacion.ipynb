```python
import sys

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "Desktop/Proyectos/spam/igj-asambleas.csv"

df = pd.read_csv(path, encoding='cp1252')
df['nombre_completo'] = df.descripcion_tipo_societario + ' ' + df.razon_social
df
```

```python
print("hay {0} un total de asambles pero {1} empresas".format(df.shape[0], len(df.nombre_completo.unique())))
```

```python
fecha_creacion = []
cantidad_asambleas = []

empresas = df.nombre_completo.unique()

for cnt, empresa in enumerate(empresas):
    sys.stdout.write('\r{0}/{1}'.format(cnt, len(empresas)))
    
    df_filtrado = df[df.nombre_completo == empresa]
    fecha_creacion.append(df_filtrado.iloc[0]['fecha_realizacion'])
    cantidad_asambleas.append(df_filtrado.shape[0])
```

```python
columns=['empresa', 'fecha_creacion', 'cantidad_asambleas']
df2 = pd.DataFrame([[empresa, fecha, asamblea] for 
                    empresa, fecha, asamblea in zip(empresas, fecha_creacion, cantidad_asambleas)], columns=columns)
df2
```

```python
sum(df2.cantidad_asambleas)
```

```python
df2.to_csv('fecha_creacion_completo.csv', index=False)
```

```python
df3 = df2[np.isfinite(df2.fecha_creacion)]
df3.fecha_creacion = df3.fecha_creacion.apply(int)
df3
```
